```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import NoAlertPresentException
import sys
import numpy as np
import unittest, time, re
from collections import Counter
import requests
from  more_itertools import unique_everseen
import numpy as np
```

```python
def process_str(x):
    title=x.split("by")[0].strip()
    views=x.split(" ")[-2]
    return title,views

def get_format_data(l):
    titles=[]
    views=[]
    for i in l:
        title,view=process_str(i)
        titles.append(title)
        views.append(view)
    return titles,views

def get_title_views(url):
    driver = webdriver.Chrome()
    driver.implicitly_wait(30)
    driver.get(url)
    last_len=0

    for i in range(1,30):
    #     driver.execute_script("window.scrollBy(0, 500);")
        driver.find_element_by_tag_name('body').send_keys(Keys.END)
        cur_len=len(driver.page_source)
        if cur_len==last_len:
            break
        last_len=cur_len
        print(cur_len)
        time.sleep(1)
    t=driver.page_source
    l=[i.group(1) for i in re.finditer(r'aria-label="([^"]+views)"',t)]
    return get_format_data(l)

def get_like_dislike(url):
    r=requests.get(url)
    t=r.text
    l,d=re.findall('likeCount\\\\\":([0-9]+)',t)
    return int(l),int(d)

def get_videos_info(url):
    driver = webdriver.Chrome()
    driver.implicitly_wait(30)
    driver.get(url)
    last_len=0

    for i in range(1,30):
    #     driver.execute_script("window.scrollBy(0, 500);")
        driver.find_element_by_tag_name('body').send_keys(Keys.END)
        cur_len=len(driver.page_source)
        if cur_len==last_len:
            break
        last_len=cur_len
        print(cur_len)
        time.sleep(1)
    t=driver.page_source
    l=[i.group(1) for i in re.finditer(r'aria-label="([^"]+views)"',t)]
    title,views=get_format_data(l)
    
    u=re.findall('(/watch\?[^"]+)"',t)
    u=unique_everseen(u)
    urls=["http://www.youtube.com"+i for i in u]
    likes,dislikes=[],[]
    c=0
    total=len(urls)
    for i in urls:
        try:
            a,b=get_like_dislike(i)
        except:
            a,b=0,0
        likes.append(a)
        dislikes.append(b)
        c+=1
        print(round(c/total,2),end="\r")
    views=[int("".join(i.split(","))) for i in views]
    return title,views,urls,likes,dislikes
    
    
    
    
```
